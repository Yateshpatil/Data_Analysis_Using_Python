#  Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
df.columns
```

```python
df["Rating"]
```

```python
def rating_func(rating):
    if rating>7:
        return "Excellent"
    elif rating<=4:
        return 'Average'
    else:
        return 'Good'
        
```

```python
df["Rating_new"]=df["Rating"].apply(rating_func)
```

#  Find Unique Values From Genre 

```python
df["Genre"]
```

```python
list1=[]
```

```python
for value in df["Genre"]:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```

```python
len(uni_list)
```
